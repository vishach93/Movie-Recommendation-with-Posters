```python
#Converting string of List into List with ast.literal_eval
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
#list of all the genres for the a movie
movies['genres'] = movies['genres'].apply(convert)
```

```python
#List of all the keywords for a movie

movies['keywords']= movies['keywords'].apply(convert)
```

```python
#function to get first 3 important actors for a movie

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
            counter += 1
        else:
            break
    return L
```

```python
#list of important actors for a movie 
movies['cast'] = movies['cast'].apply(convert3)
```

```python
#fetching director from the string list for each movie

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            L.append(i["name"])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#converting string into list

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
#Removing spaces between words

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", '') for i in x])
```

```python
#getting all the important columns into a new_df

new_df = movies[['movie_id','title', 'tags']]
```

```python
#converting the list in tags column into a string 
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
#converting into lower case
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
#stemming
from nltk.stem.porter import PorterStemmer 

# initialize object of PorterStemmer
ps = PorterStemmer()

def stem(text):
    
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
# checking stem func
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#taking 5000 words corpus 
#removing stop words in english language

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
#transforming with 5000 words
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
#stemmed
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#calculating cosine similarity
#calculating distance betweeen each movie with all other movies
#to find similarity between two movies
similarity = cosine_similarity(vectors)
```

```python
# simlilarity matrix
# array of arrays
similarity
```

```python
#function to recommend 5 similar movies

def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] #index of movie
    
    distances = similarity[movie_index] # calculate distance between each movie
    
    #0th index is self so..
    #sorting on first 5 similar movie starting from 1 
    movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:6]
    
    
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title) #printing titles of similar movies
        
        #print(i[0])    #printing indexes of similar movies
    
    
    
```

```python
# testing recommendation of 5 similar movies based on cosine similarity
recommend('Batman Begins')
```
